#### Function to create Horizontal Bar plots

```python
def plot_frequency(data, feature1, feature2, x_lbl, y_lbl, text, color):
    
    new_df = data   #creating a temporary dataframe in order to avoid making changes in the original dataframe.
    
    new = new_df.sort_values(by = [feature2], ascending = True) # sorting in descending order according to the second feature.
    
    plt.figure(figsize = (14, 8))   # define the size of the plot
    
    x_state = list(new[feature1].values) # defining x axis
    y_state = list(new[feature2].values) # defining y axis (count of numerical entity)
    
    plt.barh(x_state, y_state, color = color)   #horizontal bars
    
    plt.xlabel(x_lbl, fontsize = 18) # xlabel
    plt.ylabel(y_lbl, fontsize = 18) # ylabel
    
    plt.title(text, fontsize = 18)  #title 
    
    return plt.show()     
```

```python
plot_frequency(charging_station_highway,"Highways/Expressways","Charging Stations", "Number of Charging Stations",
               "Highway","Number of Charging Stations on Highways",'black')
```
